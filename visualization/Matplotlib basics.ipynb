## 2 plotting interfaces
- Matlab style interface: plt.* functions as demonstrated above. this interface is stateful: it keeps track of the current figure and axes, which are where all plt commands are applied. Can use plt.gcf() to get current figure and plt.gca() to get current axes.
- Object oriented interface:
 - figure object: the top level container for all the plot elements (axes), the whole figure
 - axes object: contains most of the figure elements: Axis, Tick, Line2D, Text, Polygon, etc., and sets the coordinate system
- axes plot functions is similar to plt.plot, plt.bar...
- axes setting functions typically add a set_ prefix, ax.set_title, ax.set_xlabel, ax.set_xlim

```python
# redo last figure using object-oriented interfaces
# if sharex/sharey True, only keep left y-axis or bottom x-axis
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(10, 10))
fig.subplots_adjust(hspace=0, wspace=0.1)
ax1, ax2 = axes

ax1.plot(x, np.sin(x), 'b-', label=r'$\sin(\theta)$');
ax1.plot(x, np.cos(x), 'g--', label=r'$\cos(\theta)$');
ax1.plot(x, np.sin(x)+np.cos(x), 'm-+', label=r'$\sin(\theta)+\cos(\theta)$');
ax1.set_title('trigonometric curves');
# ax1.set_xlabel(r'$\theta$');
ax1.set_ylabel(r'f($\theta$)');
# plt.axis[0, np.ceil(N*pi), -2, 2])  # [xmin, xmax, ymin, ymax]
ax1.axis('scaled')
ax1.set_xlim([0, np.ceil(N*pi)])
ax1.set_ylim([-2, 3])
ax1.grid('both')
# x_ticks = plt.xticks()  # get x_ticks object (loc, array of text objs)
x_ticks = np.arange(N+1)
x_tick_values = np.arange(N+1)*pi
x_tick_labels = [f'${val} \pi$' if val else '0' for val in x_ticks]
#ax1.set_xticks(x_tick_values)
#ax1.set_xticklabels(x_tick_labels)
ax1.set_yticklabels(ax1.get_yticks(), rotation=-30)

# use labels defined in plot function
ax1.legend(loc='upper center', ncol=3);

ax2.plot(x, np.sin(x), 'b-', label=r'$\sin(\theta)$');
ax2.plot(x, np.cos(x), 'g--', label=r'$\cos(\theta)$');
ax2.plot(x, np.sin(x)+np.cos(x), 'm-+', label=r'$\sin(\theta)+\cos(\theta)$');
#ax2.set_title('trigonometric curves');
ax2.set_xlabel(r'$\theta$');
ax2.set_ylabel(r'f($\theta$)');

ax2.axis('scaled')
ax2.set_xlim([0, np.ceil(N*pi)])
ax2.set_ylim([-2, 3])
ax2.grid('both')
# x_ticks = plt.xticks()  # get x_ticks object (loc, array of text objs)
x_ticks = np.arange(N+1)
x_tick_values = np.arange(N+1)*pi
x_tick_labels = [f'${val} \pi$' if val else '0' for val in x_ticks]
ax2.set_xticks(x_tick_values)
ax2.set_xticklabels(x_tick_labels)
ax2.set_yticklabels(ax2.get_yticks(), rotation=30)

# use labels defined in legend function
# disable outer frame of the legend
ax2.legend(['sin', 'cos', 'sin+cos'], loc='upper right', frameon=False);
```

## Annotate plots
- text
- annotate
- lines

```python
# plot 3 lines with different colors and linestyle
# add title, xlabel, and ylabel, can use latex style input
plt.plot(x, np.sin(x), 'b-', label=r'sin($\theta$)');
plt.plot(x, np.cos(x), 'g--', label=r'cos($\theta$)');
plt.title('trigonometric curves');
plt.xlabel(r'$\theta$');
plt.ylabel(r'f($\theta$)');
plt.ylim(-1, 1.67);

# add a text
plt.text(2, 1.3, 'two curves', bbox=dict(fc="pink", ec='red'));
# add an annotate
plt.annotate('max values',
             xy=(2*np.pi, 1),
             xytext=(3*np.pi, 1.3),
             arrowprops=dict(facecolor='black', arrowstyle="->",
                            connectionstyle="angle,angleA=0,angleB=90,rad=10")
            )
plt.annotate('max values',
             xy=(2.5*np.pi, 1),
             xytext=(3*np.pi, 1.3),
             arrowprops=dict(facecolor='black', arrowstyle="->")
            )
# add lines
plt.vlines(2.25*np.pi, -1, 1.5, color='y')
# add line along x/y axis
plt.axhline(y=0)

plt.legend();
```
